```python
df.groupby('Category')['Sales'].sum()
```

```python
import matplotlib.pyplot as plt

df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.show()
```

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.show()
```

```python
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
df['Postal Code'].fillna(df['Postal Code'].mean(), inplace=True)
```

```python
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Sub-Category')['Sales'].sum().sort_values().plot(kind='barh')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)

df['Month'] = df['Order Date'].dt.month

df.groupby('Month')['Sales'].sum().plot(kind='line', marker='o')
plt.show()
```

```python
df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.show()
```

```python
df.groupby('Segment')['Sales'].sum()
```

```python
df.groupby('Segment')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
import seaborn as sns

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
print("Top Category:\n", df.groupby('Category')['Sales'].sum().sort_values(ascending=False))

print("\nTop Region:\n", df.groupby('Region')['Sales'].sum().sort_values(ascending=False))

print("\nTop City:\n", df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(5))

print("\nSegment Contribution:\n", df.groupby('Segment')['Sales'].sum())

print("\nMonthly Sales:\n", df.groupby('Month')['Sales'].sum())
```
